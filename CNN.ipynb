```python
import torch
```

```python
# CNN 구현 

from torchvision import datasets
from torchvision.transforms import ToTensor
train_data = datasets.MNIST(
    root = "data",
    train = True ,
    transform = ToTensor(), # 데이터를 가져와서 텐서로 변경시켜줌
    download = True,
)
test_data = datasets.MNIST(
    root = "data",
    train = False ,
    transform = ToTensor(),
)
```

```python
# DataLoader에 입력해서 프로그램을 구성 
# row 데이터를 받아서 torch의 신경망 모델어 넣어서 연산을 수행하게끔 하게하는 코드
# 보다 더 효율적이고 빠른 training이 될 수 있도록 하는것이 DataLoader

from torch.utils.data import DataLoader
loaders = {
    'train' : torch.utils.data.DataLoader(train_data,
                                         batch_size = 100,
                                         shuffle=True,
                                         num_workers = 1),
    'test' : torch.utils.data.DataLoader(test_data,
                                        batch_size=100,
                                        shuffle = True,
                                        num_workers=1),
}
loaders
```

```python
# CNN model 정의 
# class를 선언하는것으로 시작 
class CNN(torch.nn.Module):

    # init part : 이러 이러한 레이어들을 forward에서 사용하겠다
    def __init__(self):
        super(CNN , self).__init__()
        self.layer1 = torch.nn.Sequential(
            torch.nn.Conv2d(1, 16 , kernel_size = 5 , stride = 1 , padding = 2), # 1차원 데이터(채널이 1개)로 받아서 16개로 채널로 나누겠다는 의미 , 5x5 행렬의 커널을 이용해서 16개의 feture map을 만들겠다 , stride 만큼 움직이겠다.
            # padding = 2이기 때문에 28 x 28 ==> 32 x 32 ==> 28 x 28 이 됨
            torch.nn.ReLU(), 
            torch.nn.MaxPool2d(kernel_size =2 , stride = 2) # 28 x 28 ==> 14 x 14
        )
        self.layer2 = torch.nn.Sequential(
                torch.nn.Conv2d(16, 32 , kernel_size = 5 , stride = 1 , padding = 2), # 14 x 14 ==> 18 x 18 ==> 14 x 14
                torch.nn.ReLU(),
                torch.nn.MaxPool2d(kernel_size =2 , stride = 2) # 14 x 14 => 7 x 7
        )
        self.fc = torch.nn.Linear(32 * 7 * 7 , 10 , bias = True) # feature map dim * feature size / output = 10 (class) 
        torch.nn.init.xavier_uniform_(self.fc.weight)

    # 순전파 하겠다
    def forward(self , x): # 순전파만 만들어주면 역전파는 자동적으로 연산이 실행됨
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.view(out.size(0) , -1)
        out = self.fc(out)
        return out
```

```python
model = CNN()
```

```python
model
```

```python
# 모델의 흐름을 아는것이 중요
learning_rate = 0.01
loss_func = torch.nn.CrossEntropyLoss() # 효율적인 손실함수
optimizer = torch.optim.Adam(model.parameters() , lr = learning_rate) # local minimun 문제 , 기울기 소실 등등 학습하는 과정 중에 발생하는 문제들을 어느정도 해결해주는 para
```

```python
training_epochs = 10
```

```python
total_batch = len(loaders['train'])
for epoch in range(training_epochs):
    avg_cost = 0 

    for X , Y in loaders['train']:
        optimizer.zero_grad()
        pred = model(X)
        cost = loss_func(pred , Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost / total_batch 
        
    print('[Epoch : {:>4}] cost = {: >.9}'.format(epoch + 1 , avg_cost))
    
print('learning Finished')
```

```python
# 생성된 모델의 성능을 파악
figure = plt.figure(figsize = (10 , 8))
cols , rows = 5,5
for i in range(1 , cols * rows + 1):
    sample_idx = torch.randint(len(test_data) , size = (1,)).item()
    test_output = model(torch.unsqueeze(loaders['test'].dataset[sample_idx][0] , dim = 0))
    label = torch.argmax(test_output)
    figure.add_subplot(rows , cols , i)
    plt.title(label.numpy())
    plt.axis('off')
    plt.imshow(loaders['test'].dataset[sample_idx][0].squeeze().detach().numpy() , cmap = 'gray')
plt.show()

# 상단의 title가 예측값 , 하단의 image가 실제값
```
